```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Replace with actual dataset)
df = pd.DataFrame({
    'Item Type': ['A', 'A', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'E'],
    'Total Revenue': [500, 700, 800, 650, 900, 720, 780, 850, 900, 950]
})

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Item Type', y='Total Revenue', data=df)

# Add labels and title
plt.xlabel("Item Type")
plt.ylabel("Total Revenue")
plt.title("Box Plot of Total Revenue by Item Type")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Replace with your actual dataset)
df = pd.DataFrame({
    'Unit Sold': [100, 150, 200, 250, 300],
    'Unit Price': [10, 12, 15, 14, 13],
    'Total Revenue': [1000, 1800, 3000, 3500, 3900],
    'Total Cost': [600, 1100, 2200, 2800, 3100],
    'Total Profit': [400, 700, 800, 700, 800]
})

# Compute correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix of Sales Data")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Replace 'sales_data.csv' with your actual file path)
df = pd.read_csv(r"C:\Users\user\Desktop\Group Project\Data_Science_Project_Group\1500000 Sales Records (1).csv")  # If Excel, use pd.read_excel("file.xlsx")

# Display first few rows to verify columns
print(df.head())

# Check for missing values in relevant columns
df = df[['Sales Channel', 'Total Profit']].dropna()

# Create a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Sales Channel', y='Total Profit', data=df, palette="muted")

# Add labels and title
plt.xlabel("Sales Channel")
plt.ylabel("Total Profit")
plt.title("Violin Plot of Total Profit by Sales Channel")

# Show the plot
plt.show()
```
