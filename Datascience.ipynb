```python
import pandas as pd

file_path = ("C:/Users/user/Desktop/Machine_learning/Data_Science_Project_Group/1500000 Sales Records (1).csv")
df = pd.read_csv(file_path)

df.head(), df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.violinplot(x="Sales Channel", y="Total Profit", data=df, ax=axes[0], palette="muted")
axes[0].set_title("Total Profit by Sales Channel")

sns.violinplot(x="Sales Channel", y="Total Revenue", data=df, ax=axes[1], palette="pastel")
axes[1].set_title("Total Revenue by Sales Channel")

plt.tight_layout()
plt.show()
```
