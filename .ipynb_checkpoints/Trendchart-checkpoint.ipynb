```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\ADMIN\Desktop\KeSRA Data Science Training\Week 5\GIT Hub\Week 1 practice\Project_Data_Collaboration\Data_Science_Project_Group\1500000 Sales Records (1).csv"  # Update path accordingly

df = pd.read_csv(file_path)
```

## Trend chart of total profit by order date

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

plt.figure(figsize=(10, 6))

df.groupby(df['Order Date'].dt.year)['Total Profit'].sum().plot(kind='line', marker='o', color='green')

plt.title('Total Profit Over the Years')

plt.xlabel('Year')

plt.ylabel('Total Profit')

plt.grid(True)

plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

plt.figure(figsize=(10, 6))

df.groupby(df['Order Date'].dt.year)['Units Sold'].sum().plot(kind='line', marker='o', color='green')

plt.title('Total Units Sold Over the Years')

plt.xlabel('Year')

plt.ylabel('Total Units Sold')

plt.grid(True)

plt.show()
```
