### Distribution of revenue by order priority (bar chart and pie chart)

```python
import pandas as pd
df = pd.read_csv("sales_records.csv")
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate revenue by Order Priority
df_grouped = df.groupby("Order Priority")["Total Revenue"].sum().reset_index()

# Sort by revenue for better visualization
df_grouped = df_grouped.sort_values(by="Total Revenue", ascending=False)

# Plot Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x="Order Priority", y="Total Revenue", data=df_grouped, palette="viridis")

# Labels and Title
plt.xlabel("Order Priority")
plt.ylabel("Total Revenue")
plt.title("Revenue Distribution by Order Priority")
plt.xticks(rotation=45)  # Rotate labels if necessary

# Show the plot
plt.show()
```

```python
labels = df_grouped["Order Priority"]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

plt.figure(figsize=(4, 4))
plt.pie(df_grouped["Total Revenue"], labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})

# Title
plt.title("Revenue Distribution by Order Priority")

# Show Plot
plt.show()
```
