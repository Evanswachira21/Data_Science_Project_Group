```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("1500000 Sales Records (1).csv")
df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Units Sold'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Total Units sold")
plt.ylabel("Frequency")
plt.title("Histogram of Total Units Sold")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python

sns.histplot(df["Units Sold"], bins=20, kde=True)

plt.title("Distribution of Units Sold")

plt.show()
```

```python
#Violin Total profits and sales channel
plt.figure(figsize=(6, 5))
sns.violinplot(x='Sales Channel', y='Total Profit', data=df,palette={"Online": "blue", "Offline": "purple"})
# Labels and title
plt.xlabel('Sales Channel')
plt.ylabel('Total Profit')
plt.title('Violin Plot of Total Profit by Sales Channel')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.barplot(x="Item Type", y="Total Revenue", data=df, estimator=sum, palette="magma")
plt.xlabel("Item Type")
plt.ylabel("Total Revenue ($)")
plt.title("Total Revenue per Item")
plt.show()


```

```python
df = pd.DataFrame(df)


df_grouped = df.groupby("Order Priority")["Total Revenue"].sum().reset_index()
priority_colors = {"C": "blue", "H": "purple", "L": "skyblue", "M": "orange"}
plt.figure(figsize=(8, 5))
plt.bar(df_grouped["Order Priority"], df_grouped["Total Revenue"],
       color=[priority_colors[p] for p in df_grouped["Order Priority"]])
plt.xlabel("Order Priority")
plt.ylabel("Total Revenue ($)")
plt.title("Total Revenue per Order of Priority")
plt.show()
```
